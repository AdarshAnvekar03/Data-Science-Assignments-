```python
movie_data=movie_data.fillna(value=0)
```

```python
movie_cleaned=movie_data.iloc[:,5:15]
movie_cleaned
```

```python
movie_cleaned.head()
```

```python
freq_itemsets = apriori(df =movie_cleaned,min_support=0.02,use_colnames=True) #apriori(df,min_support=0.5,)
freq_itemsets.head(20)
```

```python
best_assosiates = association_rules(df = freq_itemsets,metric='confidence',min_threshold=0.70)
best_assosiates
```
