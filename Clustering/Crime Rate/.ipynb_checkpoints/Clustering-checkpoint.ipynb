```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.preprocessing import LabelEncoder
lab_en=LabelEncoder()
from sklearn.cluster import KMeans,DBSCAN,AgglomerativeClustering
#plt.style.use("seaborn-darkgrid")
import warnings
warnings.filterwarnings("ignore")
```

```python
crime_data=pd.read_csv("crime_data.csv")
crime_data.head()
```

```python
crime_data.shape
```

```python
crime_data.dtypes
```

```python
crime_data.isna().sum()
```

```python
crime_data.rename(columns={'Unnamed: 0' : "unnamed"},inplace=True)
crime_data.head()
```

```python
crime_data.head(2)
```

```python
del crime_data['unnamed']
```

```python
crime_data.head()
```

```python
for i in range(1,10):
    kmean=KMeans(n_clusters=i)
    kmean.fit(crime_data)
    pred = kmean.predict(crime_data)
    plt.scatter(crime_data["Assault"],crime_data["Murder"],c=pred,cmap=plt.cm.Accent)
        
    plt.show()
```

```python
pred=kmean.predict(crime_data)
pred
```

### DB Scan

```python
db=DBSCAN(eps=21,min_samples=4)
pred=db.fit_predict(crime_data)
plt.scatter(crime_data['Assault'],crime_data['UrbanPop'],c=pred)
```

```python
db.fit_predict(crime_data)
```

### Hierarchical Clustering

```python
hie_clu=AgglomerativeClustering(n_clusters=4)
pred=hie_clu.fit_predict(crime_data)
plt.scatter(crime_data['Assault'],crime_data['UrbanPop'],c=pred)
```

```python
for i in range(1,16):
    hie_clu=AgglomerativeClustering(n_clusters=i)
    pred=hie_clu.fit_predict(crime_data)
    plt.scatter(crime_data['Assault'],crime_data['UrbanPop'],c=pred)
    plt.show()
```
