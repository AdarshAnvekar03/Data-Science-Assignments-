```python
import pandas as pd
import seaborn as sns 
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans,DBSCAN,AgglomerativeClustering
#plt.style.use("seaborn-darkgrid")
import warnings
warnings.filterwarnings("ignore")
```

```python
airline_data=pd.read_excel("EastWestAirlines.xlsx",sheet_name="data")
airline_data
```

### Kmeans

```python
%%time
for i in range(1,10):
    kmean=KMeans(n_clusters=i)
    kmean.fit(airline_data)
    pred = kmean.predict(airline_data)
    plt.scatter(airline_data['Days_since_enroll'],airline_data['Qual_miles'],c=pred,cmap=plt.cm.Accent)
        
    plt.show()
```

```python
kmean=KMeans(n_clusters=10)
kmean.fit(airline_data)
pred = kmean.predict(airline_data)
plt.scatter(airline_data['Days_since_enroll'],airline_data['Qual_miles'],c=pred)
        
```

### DBSCAN

```python
mean=DBSCAN(eps=21,min_samples=1)
pred =mean.fit_predict(airline_data)
plt.scatter(airline_data['Days_since_enroll'],airline_data['Qual_miles'],c=pred)
        
```

### Hierarchical Clustering

```python
hie_clu=AgglomerativeClustering(n_clusters=4)
pred=hie_clu.fit_predict(airline_data)
plt.scatter(airline_data['Days_since_enroll'],airline_data['Qual_miles'],c=pred)
```

```python
for i in range(1,16):
    hie_clu=AgglomerativeClustering(n_clusters=i)
    pred=hie_clu.fit_predict(airline_data)
    plt.scatter(airline_data['Days_since_enroll'],airline_data['Qual_miles'],c=pred)
    plt.show()
```
