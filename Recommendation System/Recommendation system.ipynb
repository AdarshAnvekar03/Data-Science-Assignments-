```python
corr_piv= pd.pivot_table(data =book_data,values='Book.Rating',index='User.ID',columns='Book.Title').fillna(0)
corr_piv
```

```python
corr_piv.values
```

```python
cosine_ubcf = 1 - pairwise_distances(X = corr_piv.values,metric='cosine')
cosine_ubcf
```

```python
cosine_ubcf_df         = pd.DataFrame(data=cosine_ubcf)
cosine_ubcf_df.index   = book_data['User.ID'].unique()
cosine_ubcf_df.columns = book_data['User.ID'].unique()
cosine_ubcf_df.round(2)
```

```python
Top_50data=cosine_ubcf_df.iloc[:50,:50]
Top_50data.idxmax()
```
