```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
lab_en=LabelEncoder()
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import GaussianNB
gauss=GaussianNB()
multi=MultinomialNB()
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
salary_data=pd.read_csv("SalaryData_Train.csv")
salary_data
```

```python
salary_data.shape
```

```python
salary_data.dtypes
```

```python
salary_data.isna().sum()
```

```python
salary_data["workclass"]=lab_en.fit_transform(salary_data["workclass"])
salary_data["education"]=lab_en.fit_transform(salary_data["education"])
salary_data["maritalstatus"]=lab_en.fit_transform(salary_data["maritalstatus"])
salary_data["occupation"]=lab_en.fit_transform(salary_data["occupation"])
salary_data["relationship"]=lab_en.fit_transform(salary_data["relationship"])
salary_data["race"]=lab_en.fit_transform(salary_data["race"])
salary_data["sex"]=lab_en.fit_transform(salary_data["sex"])
salary_data["native"]=lab_en.fit_transform(salary_data["native"])
salary_data["Salary"]=lab_en.fit_transform(salary_data["Salary"])
```

```python
salary_data.head()
```

```python
X=salary_data.drop(labels="Salary",axis=1)
y=salary_data[['Salary']]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=12,)
```

```python
y_pred_gtest=gauss.fit(X_train,y_train).predict(X_test)
y_pred_mtest=multi.fit(X_train,y_train).predict(X_test)
```

### Test for Gaussian 

```python
accuracy_score(y_test,y_pred_gtest)
```

```python
confusion_matrix(y_test,y_pred_gtest)
```

### Test for multinomial

```python
accuracy_score(y_test,y_pred_mtest)
```

```python
confusion_matrix(y_test,y_pred_mtest)
```
